```python
import pandas as pd
```

```python
"""La base de datos de personas registradas"""

#Los datos deben estar ordenados por fecha más reciente de forma descendente 
#Único dato variable
DB = pd.read_csv('/content/AJEDREZ UVG - Registro-6.csv')
#Selecciona las columnas requeridas 
DB_columns= DB.loc[0:len(DB),["¿Nombre y apellido?","¿Carrera? ","Usuario"]]
#Elimina a las personas que no tienen usuarios
DB_columns= DB_columns[DB_columns['Usuario'].notna()]
#Convierte la columna de usuario a minúsculas
DB_columns["Usuario"]= DB_columns["Usuario"].str.lower()
#Se limpian los usuarios duplicados y se mantiene el dato más reciente.
DB_limpia = DB_columns.drop_duplicates(subset=['Usuario'])
DB_limpia.columns = ['Nombre', "Carrera", "Usuario"]
DB = DB_limpia
```

```python
"""Función para ordenadar los datos de chess.com"""
def Ronda_limpia(path, ronda): 
  Ronda_chess = pd.read_csv(path)
  Ronda= Ronda_chess.loc[0:len(Ronda_chess), ["Username","Score","SB"]]
  #Convierte la columna de usuario a minúsculas
  Ronda["Username"]= Ronda["Username"].str.lower()
  #Reemplaza la anterior columna de usuarios a la nueva
  Ronda_final = Ronda.assign(Username= Ronda["Username"])
  Ronda_final.columns = ['Usuario', 'R'+str(ronda), 'D'+str(ronda)]
  return Ronda_final

#Se puede a consultar a más detalle el uso de "outer" y "left" en https://stackoverflow.com/questions/53645882/pandas-merging-101

def Juntar_Rondas_con_DB(rondas):
  return pd.merge(DB, rondas, on="Usuario", how= "right")

def Rondas_Individuales(rondaR,rondaL):
  return pd.merge(Ronda1,Ronda2, on="Usuario", how="outer")
```
